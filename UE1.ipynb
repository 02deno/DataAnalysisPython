## Aufgabe 1
Überprüfen Sie die Datei und den Dataframe. Die Daten und Zeiten sind in separaten Spalten angegeben. Wir werden sie nun in eine einzige Datums- und Zeitspalte im generischen Format "parsen". Was ist die Fehlerquelle? Beheben Sie den Fehler durch eine bedingte Änderung des Arrays. 

Inspect the file and the dataframe. The dates and times are given in separate columns. We are now going to parse them into a single datetime column in generic format. What is the source of the error? Fix the error, by conditional array modification. 

```python
import pandas as pd
import numpy as np
import datetime as dt


sunspot = pd.read_csv("SunspotGroupData.txt",sep='\s+')
sunspot["se"].replace({60: 59}, inplace=True)
sunspot['date'] = sunspot['yr'].astype(str) + '-' + sunspot['mo'].astype(str) + '-' + sunspot['da'].astype(str) + ' ' + sunspot['hr'].astype(str) + ':' + sunspot['mi'].astype(str)+ ':' + sunspot['se'].astype(str)
sunspot['date'] = pd.to_datetime(sunspot.date, format='%Y-%m-%d %H:%M:%S')
sunspot
#sunspot.date.dtype
```

Now we are converting the date-times into decimal form, using the 'dayofyear' attribute of datetime. 

```python
sunspot.date.dt.year + sunspot.date.dt.dayofyear/365.25 + sunspot.date.dt.hour / 24 + sunspot.date.dt.minute / (24*60)+ sunspot.date.dt.second / (24*60*60)
```

It is always good to copy the first two rows of the data file onto your notebook, to see which features we have. Here,  
**i**: Means that this data belong to sunspot *groups*, not to individual sunspots  
**yr, mo, da, hr, mi, se**: year, month, day, hour, minute, second  
**id**: ID number of the sunspot group  
**pr1, pr2**: Projected group-area quantities. Just ignore them for this task  
**ua**: Umbral area of sunspot group (umbra is the darkest central region of a sunspot)  
**ws**: Whole-spot area of sunspot group (in millionths of the visible solar hemispheric area)  
**lat**: Latitude (degrees)  
**lon**: Longitude (degrees)  
**lcm**: Longitudinal distance from the Sun's central meridian  
**pa**: Position angle (irrelevant now)  
**dcen**: Distance from the centre of Sun's disc measured in units of the solar radius

## Aufgabe 2
Finden und eliminieren Sie Zeilen mit fehlenden Daten. Tun Sie dies in dem Array oder Dataframe, den Sie oben erstellt haben.

Find and eliminate rows with missing data. Do this in the array or dataframe you created above. 

```python
sunspot = sunspot[sunspot.lat != 999999]
sunspot
```

Um Messfehler zu reduzieren, filtern wir die Sonnenfleckengruppen innerhalb eines 2x70-Grad-Längsfensters um den Zentralmeridian. 

To reduce measurement errors, we are filtering those sunspot groups within a 2x70-degree longitudinal window around the central meridian. 

```python
sunspot = sunspot[np.abs(sunspot.lcm) <= 70]
sunspot
```

Um das durch die kleinsten Fleckengruppen verursachte Rauschen zu reduzieren, filtern wir die Daten nur nach Flächenwerten, die nicht Null sind. 

To reduce noise due to the smallest-area groups, we are filtering the data for non-zero area values only. 

```python
sunspot = sunspot[sunspot.ws !=0]
sunspot
```
